```python
import matplotlib.pyplot as plt
```

```python
def labels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i]) #(posx , posy , valor a mostrar)
```

```python
x=['Ringo', 'John', 'Paul', 'George', 'Yoko']
y=[45, 34, 42, 38, 47]
plt.figure(figsize=(10,5))
plt.bar(x,y)
labels(x,y) #Definida por nosotros
plt.title("Titulo de figura")
plt.xlabel("Texto eje X")
plt.ylabel("Texto Eje Y")
plt.xticks(rotation=65)
plt.show() #Muestra la gráfica
```
